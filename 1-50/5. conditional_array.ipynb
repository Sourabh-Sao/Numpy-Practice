```python
import numpy as np
```

### Given two arrays:

a = np.array([10, 20, 30, 40, 50])

b = np.array([5, 25, 15, 45, 35])

Create a new array where:

If a[i] > b[i] → put a[i]*2

Else → put b[i]-10

Do not use loop, use np.where

```python
a = np.array([10, 20, 30, 40, 50])
b = np.array([5, 25, 15, 45, 35])

c=np.where(a>b,a*2,b-10)
c
```

### The "Scalable" Approach

```python
# If you ever have more than two conditions
condition=[a>b,a<=b]
choice=[a*2,b-10]

c=np.select(condition,choice)
c
```
