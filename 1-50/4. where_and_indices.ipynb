```python
import numpy as np
```

### Given this array:

scores = np.array([45, 82, 19, 67, 91, 33, 58, 76])

Find and print:

- All scores >= 60
- How many students passed (>=60)
- Positions (indices) of students who scored less than 40

```python
scores = np.array([45, 82, 19, 67, 91, 33, 58, 76])

# this is boolean condition
a=scores>=60
print('scores greater and equal to 60 : ',scores[a])
# scores[a]
```

```python
scores = np.array([45, 82, 19, 67, 91, 33, 58, 76])

# a=scores>=60
# len(scores[a])

# or 

a=np.where(scores>=60)
len(scores[a])
```

```python
scores = np.array([45, 82, 19, 67, 91, 33, 58, 76])
a=np.where(scores<40)
a
```
