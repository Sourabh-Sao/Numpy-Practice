Sort the names according to descending order of marks (without using sorted() or pandas).

Print names in that sorted order.

Hint: Use np.argsort() with trick.

```python
names = np.array(["Rahul", "Priya", "Aman", "Sneha", "Vikram"])

marks = np.array([78, 92, 65, 88, 71])

# Returns the indices that would sort an array.
desc=np.argsort(marks)[::-1] # descending order
asc=np.argsort(marks) # by-default ascending 

desc2=np.argsort(-marks) # it also sort indices in descending
names[desc2]

```
