### Convert it into 2D array of shape (3,2) in two different ways:

arr = np.array([1, 2, 3, 4, 5, 6])

- Using reshape
- Using np.newaxis or expand_dims

Both should give same result.

```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr1=np.reshape(arr,(3,2),order='A')

arr1
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr_copy=arr.copy()
arr_copy.resize((3,2))

arr_copy
```
