### Given time series array (daily temperatures):

temps = np.array([22.5, 23.1, 24.0, 23.8, 25.2, 26.1, 25.9, 24.5, 23.0])

- Find day-to-day differences (use np.diff)
- Find 25th, 50th (median), and 75th percentile of the differences
- Also find how many days temperature increased strictly (>0 diff)

```python
temps = np.array([22.5, 23.1, 24.0, 23.8, 25.2, 26.1, 25.9, 24.5, 23.0])
```

```python
diff=np.diff(temps)
diff
```

```python
a=np.percentile(diff,25)
b=np.percentile(diff,50)
c=np.percentile(diff,75)

a,b,c
```

```python
np.sum(diff>0)
```
