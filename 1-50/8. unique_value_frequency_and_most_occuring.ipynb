### Given messy array:

data = np.array([22, 15, 22, 17, 15, 30, 22, 17, 40, 15, 30])

Print:

- Unique values
- How many times each value appears (in sorted order of values)
- The value that appeared most times

(Use np.unique() with return_counts)

```python
data = np.array([22, 15, 22, 17, 15, 30, 22, 17, 40, 15, 30])

unique_values=np.unique(data)
unique_values
```

### I use zip()

```python
# how many time each value appear
data = np.array([22, 15, 22, 17, 15, 30, 22, 17, 40, 15, 30])

u_value,cnt=np.unique(data,return_counts=True)

# list(zip(u_value,cnt))

for u_value,cnt in zip(u_value,cnt):
    print(u_value,"occurs : ",cnt)
```

```python
# values that appear most times

data = np.array([22, 15, 22, 17, 15, 30, 22, 17, 40, 15, 30])

value,cnt=np.unique(data,return_counts=True)


# np.argmax(cnt) # return index of largest value
# value[np.argmax(cnt)] 

# or

value[cnt==cnt.max()] 

```
