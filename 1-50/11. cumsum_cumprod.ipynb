### Given 2D array:

sales = np.array([
    [120, 150, 180],
    [200, 220, 190],
    [110, 130, 140],
    [250, 270, 300]
])

- Compute cumulative sum along each column (axis=0) → running total per product over days
- Compute cumulative product along each row (axis=1) → growth factor if multiplying daily multipliers

```python
sales = np.array([
    [120, 150, 180],
    [200, 220, 190],
    [110, 130, 140],
    [250, 270, 300]
])
```

```python
sales.cumsum(axis=0)
```

```python
sales.cumprod(axis=1)
```
